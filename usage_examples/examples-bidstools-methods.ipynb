## Testing Methods in the `BIDSTOOLS` Module

This Jupyter Notebook serves as a platform to rigorously test the various methods included in the Python module `BIDSTOOLS`.

This module is particularly valuable for neuroimaging researchers working with BIDS-compliant
datasets, parsing and manipulating BIDS filenames and entity dictionaries, organizing large-scale
neuroimaging studies, copying or restructuring BIDS datasets and their derivatives, and generating
overview tables and tree visualizations of dataset contents.

The tests are organized into the following sections for clarity and ease of navigation:


---

##### 1.1. str2entity: Convert a BIDS-formatted string into an entity dictionary.
---

```python
######################## Testing str2entity ###################################

import clabtoolkit.bidstools as cltbids

print('Test 1: Parse a standard T1w filename with multiple entities:')
filename = 'sub-01_ses-M00_acq-3T_dir-AP_run-01_T1w.nii.gz'
entity = cltbids.str2entity(filename)
print(entity)
print('')

print('Test 2: Parse a bold functional filename:')
filename = 'sub-02_ses-baseline_task-rest_run-01_bold.nii.gz'
entity = cltbids.str2entity(filename)
print(entity)
print('')

print('Test 3: Parse a DWI filename without a run entity:')
filename = 'sub-03_ses-M06_acq-b1000_dir-PA_dwi.nii.gz'
entity = cltbids.str2entity(filename)
print(entity)
print('')

print('Test 4: Parse a derivatives filename with a desc entity:')
filename = 'sub-01_ses-M00_desc-preproc_T1w.nii.gz'
entity = cltbids.str2entity(filename)
print(entity)
print('')

print('Test 5: Parse a filename without an extension (bare suffix only):')
filename = 'sub-01_ses-M00_T1w'
entity = cltbids.str2entity(filename)
print(entity)
print('')
```

<a id="bidstools-entity2str"></a>
##### 1.2. entity2str: Convert an entity dictionary back to a BIDS-formatted string.
---

```python
######################## Testing entity2str ###################################

import clabtoolkit.bidstools as cltbids

print('Test 1: Reconstruct a T1w filename from a full entity dictionary:')
entity = {'sub': '01', 'ses': 'M00', 'acq': '3T', 'dir': 'AP', 'run': '01',
        'suffix': 'T1w', 'extension': 'nii.gz'}
result = cltbids.entity2str(entity)
print(result)
print('')

print('Test 2: Reconstruct a bold filename from a dictionary without extension:')
entity = {'sub': '02', 'ses': 'baseline', 'task': 'rest', 'run': '01', 'suffix': 'bold'}
result = cltbids.entity2str(entity)
print(result)
print('')

print('Test 3: Verify round-trip consistency (str2entity -> entity2str):')
original = 'sub-03_ses-M06_acq-b1000_dir-PA_dwi.nii.gz'
entity = cltbids.str2entity(original)
reconstructed = cltbids.entity2str(entity)
print(f'  Original:      {original}')
print(f'  Reconstructed: {reconstructed}')
print(f'  Match: {original == reconstructed}')
print('')
```

<a id="bidstools-recursively_replace_entity_value"></a>
##### 1.7. recursively_replace_entity_value: Replace entity values across all files and folders in a BIDS dataset.
---

```python
######################## Testing recursively_replace_entity_value ###################################
# NOTE: This function modifies files and folders on disk.
# Set bids_dir to a real (ideally a copy of a) BIDS dataset before running.

import clabtoolkit.bidstools as cltbids

bids_dir = '/path/bids'   # <-- Replace with your BIDS root directory

print('Test 1: Replace ses-M00 with ses-BL across all BIDS files and folders:')
cltbids.recursively_replace_entity_value(
    root_dir=bids_dir,
    dict2old={'ses': 'M00'},  # <-- This will replace both BL and M00 with V1
    dict2new={'ses': 'BL'}
)
print('Done.')
print('')

print('Test 2: Replace both subject and session values using BIDS strings as input:')
cltbids.recursively_replace_entity_value(
    root_dir=bids_dir,
    dict2old='sub-01_ses-BL',
    dict2new='sub-001_ses-M00'
)
print('Done.')
print('')

print('Test 3: Replace only the files that contain the session and acquisition equal to M00 and 3T respectively:')
cltbids.recursively_replace_entity_value(
    root_dir=bids_dir,
    dict2old={'ses': 'M00', 'acq': '3T'},
    dict2new={'ses': 'V1', 'acq': '7T'}
)
print('Done.')

```

```python
######################## Testing validate_bids_structure ###################################

import clabtoolkit.bidstools as cltbids

bids_dir = '/path/to/bids/dataset'   # <-- Replace with your BIDS root directory

print('Test 1: Run basic BIDS validation and inspect warnings:')
warnings = cltbids.validate_bids_structure(bids_dir)

if warnings:
    print('Validation notes/warnings:')
    for w in warnings:
        print(f'  ⚠️  {w}')
else:
    print('✅ Basic BIDS structure looks good!')
print('')

print('Test 2: Validate a path that does not exist (expects FileNotFoundError):')
try:
    cltbids.validate_bids_structure('/non/existent/path')
except FileNotFoundError as e:
    print(f'Caught expected error: {e}')
print('')

print('Test 3: Validate a path that is a file rather than a directory (expects NotADirectoryError):')
try:
    cltbids.validate_bids_structure('/tmp')
    # /tmp is valid, but if we pass a file it should raise — use a known file for this test
    import tempfile
    with tempfile.NamedTemporaryFile(suffix='.txt', delete=False) as f:
        tmp_file = f.name
    cltbids.validate_bids_structure(tmp_file)
except NotADirectoryError as e:
    print(f'Caught expected error: {e}')
print('')
```
