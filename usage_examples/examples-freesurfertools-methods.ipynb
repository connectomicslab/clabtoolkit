```python
######################## Testing create_from_data ###################################
import clabtoolkit.freesurfertools as cltfree
import nibabel as nib
import numpy as np
import os

fs_dir = '/opt/freesurfer/subjects'
fs_id  = 'fsaverage'
hemi   = 'lh'
annot_file = os.path.join(fs_dir, fs_id, 'label', f'{hemi}.aparc.annot')

# Load raw data from an existing annotation using nibabel
codes, regtable, regnames = nib.freesurfer.read_annot(annot_file, orig_ids=True)
regnames = [r.decode('utf-8') for r in regnames]

# Test 1: Create a new AnnotParcellation from the raw arrays
print('Test 1: Creating AnnotParcellation from raw codes, regtable and regnames...')
parc_new = cltfree.AnnotParcellation()
parc_new.create_from_data(
    codes=codes,
    regtable=regtable,
    regnames=regnames,
    annot_id='ManualDesikan',
    hemi='lh'
)
print(f'  ID       : {parc_new.id}')
print(f'  Hemi     : {parc_new.hemi}')
print(f'  Regions  : {len(parc_new.regnames)}')
print(f'  Loaded   : {parc_new.is_loaded()}')
print('')

# Test 2: Create with a custom filename association
print('Test 2: Creating AnnotParcellation with an associated filename...')
parc_fn = cltfree.AnnotParcellation()
parc_fn.create_from_data(
    codes=codes,
    regtable=regtable,
    regnames=regnames,
    annot_id='ManualDesikanWithFile',
    hemi='lh',
    filename='/tmp/lh.manual_desikan.annot'
)
parc_fn.get_info()

```

* 1.7. map_values: Map regional values to vertex-wise values
---

```python
######################## Testing map_values ###################################
import clabtoolkit.freesurfertools as cltfree
import numpy as np
import pandas as pd
import os

fs_dir = '/opt/freesurfer/subjects'
fs_id  = 'fsaverage'
hemi   = 'lh'
annot_file = os.path.join(fs_dir, fs_id, 'label', f'{hemi}.aparc.annot')

parc      = cltfree.AnnotParcellation(parc_file=annot_file, annot_id='Desikan')
n_regions = len(parc.regtable)
np.random.seed(42)

# Test 1: Map a 1D numpy array of regional values to vertices
print('Test 1: Mapping a 1D numpy array of regional thickness values to vertices...')
thickness_values = np.random.uniform(1.5, 4.5, size=n_regions)
vertex_vals = parc.map_values(thickness_values)
print(f'  Input shape : {thickness_values.shape}')
print(f'  Output shape: {vertex_vals.shape}')
print(f'  Value range : {vertex_vals.min():.2f} - {vertex_vals.max():.2f}')
print('')

# Test 2: Map a 2D numpy array (multiple measures)
print('Test 2: Mapping a 2D numpy array (thickness + curvature) to vertices...')
multi_values = np.column_stack([
    np.random.uniform(1.5, 4.5, size=n_regions),
    np.random.uniform(-1.0, 1.0, size=n_regions)
])
vertex_multi = parc.map_values(multi_values)
print(f'  Input shape : {multi_values.shape}')
print(f'  Output shape: {vertex_multi.shape}')
print('')

# Test 3: Map a pandas DataFrame (preserves column names in output)
print('Test 3: Mapping a DataFrame of regional measures to vertices...')
df_vals = pd.DataFrame({
    'thickness': np.random.uniform(1.5, 4.5, size=n_regions),
    'area'     : np.random.uniform(500, 3000, size=n_regions),
})
vertex_df = parc.map_values(df_vals)
print(f'  Output type : {type(vertex_df).__name__}')
print(f'  Output shape: {vertex_df.shape}')
print(f'  Columns kept: {list(vertex_df.columns)}')
print('')

# Test 4: Map from a CSV file
print('Test 4: Mapping regional values from a CSV file...')
csv_path = '/tmp/region_values.txt'
np.savetxt(csv_path, thickness_values)
vertex_from_file = parc.map_values(csv_path)
print(f'  Output shape: {vertex_from_file.shape}')
print('')
```
