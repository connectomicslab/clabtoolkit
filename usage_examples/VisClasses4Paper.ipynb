```python
# Importing necessary modules
import clabtoolkit.surfacetools as cltsurf
import clabtoolkit.freesurfertools as cltfree
import clabtoolkit.visualizationtools as cltvis

surf_file = "/opt/freesurfer/subjects/fsaverage/surf/lh.inflated"
annot_file = "/opt/freesurfer/subjects/fsaverage/label/lh.aparc.annot"
label_file = "/opt/freesurfer/subjects/fsaverage/label/lh.cortex.label"

# Loading the surface and annotation
surf = cltsurf.Surface(surf_file)
annot_obj = cltfree.AnnotParcellation(parc_file=annot_file)

# Loading the original annotation onto the surface
surf.load_annotation(annot_obj, parc_name="Desikan")

# Correcting the annotation
annot_obj.correct_vertexwise_parcellation(surf_file, 
                                        cortex_file=label_file, 
                                        min_cluster_size=100)
surf.load_annotation(annot_obj, parc_name="Desikan_Corrected")

# Visualize the original and corrected annotations side by side
print('Visualizing original and corrected annotations...')
plotter = cltvis.BrainPlotter()
plotter.plot(
    objs2plot=surf,
    views='medial',
    map_names=["Desikan","Desikan_Corrected"],
)
```

```python
# Importing necessary modules
import clabtoolkit.tracttools as clttract
import clabtoolkit.visualizationtools as cltvis 

# Define file paths
in_trk = "/opt/freesurfer/trctrain/hcp/mni/cc.bodyc.display.trk"
fa_image = "/opt/fsl/data/standard/FMRIB58_FA_1mm.nii.gz"

# Load the tractogram 
tract_obj_orig = clttract.Tractogram(in_trk)
tract_obj = clttract.Tractogram(in_trk)

# Interpolate the FA values onto the tractogram
tract_obj_orig.interpolate_on_tractogram(fa_image, map_name="fa")
tract_obj.interpolate_on_tractogram(fa_image, map_name="fa")

# Filter one tractogram based on streamline length
tract_obj.filter(condition="125 < length < 135")

# Visualize the original and filtered tractograms side by side
plotter = cltvis.BrainPlotter()
plotter.plot(objs2plot=[tract_obj_orig, tract_obj],map_names="fa")
```
